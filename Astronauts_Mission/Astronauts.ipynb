Data Preprocessing

```python
astronauts_df.isna().sum()
```

```python
# Drop the Nan values since it is a negligible number of records 

astronauts_df = astronauts_df.dropna()
astronauts_df.isna().sum()
```

```python
astronauts_df = astronauts_df.drop(columns = ['id', 'original_name'], axis = 1)
astronauts_df.head()
```

```python
astronauts_df['occupation'].value_counts()
```

```python
astronauts_df['occupation'] = astronauts_df['occupation'].str.lower()
astronauts_df['occupation'] = np.where((astronauts_df['occupation'] == 'space tourist'),
                                      'other (space tourist)',
                                       astronauts_df['occupation'])
astronauts_df['occupation'].value_counts()
```

Exploratory Data Analysis

```python
unique = astronauts_df.groupby('sex')['name'].nunique()
plt.gcf().set_size_inches(12,7)
ax = sns.barplot(data = unique.to_frame().reset_index(), x = 'sex', y = 'name').set(title = 'Unique Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(data = astronauts_df, y = 'occupation', orient = "h").set(
    title = 'Occupation by Count')
plt.gcf().set_size_inches(12,7)
```

```python
yearly_missions = astronauts_df.drop_duplicates(subset = 'mission_title')['year_of_mission'].value_counts().sort_index()
sns.lineplot(data = yearly_missions, x = yearly_missions.index, y = yearly_missions.values).set(
    title = 'Total Number of Missions per Year',
    xlabel = 'Year of Mission',
    ylabel = 'Number of missions');
plt.gcf().set_size_inches(15,8)
```

```python
mil_by_civ  = astronauts_df.groupby(['military_civilian']).size().reset_index(name="total_num")
plt.pie(mil_by_civ['total_num'], labels=mil_by_civ['military_civilian'], autopct='%1.1f%%', pctdistance=0.5, explode = (0.1, 0.1), colors = ['c', 'b'])
plt.title('Military by Civilian Count')
plt.gcf().set_size_inches(12,7)
```

```python
astronauts_df['age'] = astronauts_df['year_of_mission'] - astronauts_df['year_of_birth']
unique_ages = astronauts_df['age'].nunique()
age_by_nationality = astronauts_df.groupby(by = [pd.cut(astronauts_df['age'], bins = [0,30,40,50,100]), 'nationality'])['sex'].count().to_frame().reset_index()
sns.barplot(data = age_by_nationality, y = 'nationality', x = 'sex', hue = 'age', orient = 'h')
plt.gcf().set_size_inches(15,8)
plt.title('Nationality By Age Count')
plt.xlabel('Count')
plt.ylabel('Nationality');
```

```python
mission_num = astronauts_df['mission_title'].nunique()
print(f'The astronauts went on {mission_num} mission.')
```

```python
top_5_occ = astronauts_df.occupation.value_counts()[0:5]
top_5_order = ['msp', 'commander', 'pilot', 'flight engineer','psp']
plt.figure(figsize = (12,7))
pd.crosstab(astronauts_df.occupation, astronauts_df.sex).loc[top_5_order].plot(kind = 'bar')
plt.title('Top 5 Occupation by Gender')
plt.xlabel('Occupation')
plt.ylabel('Total Number');
```

```python
nationality_count = (astronauts_df['nationality'].value_counts().to_frame()).reset_index()
nationality_count.columns = ['nationality','freq']
plt.figure(figsize = (15,8))
plt.barh(nationality_count.nationality[0:10], nationality_count.freq[0:10])
plt.title('Top 10 Astronauts by Nationality')
plt.xlabel('Count')
plt.ylabel('Nationality');
```

```python
plt.figure(figsize=(15,7))
astronauts_df['total_hrs_sum'] = astronauts_df['total_hrs_sum'].astype(float)
astronauts_df['total_hrs_sum'].groupby(astronauts_df['nationality']).mean().sort_values(ascending = False)[0:10].plot(kind='bar')
plt.title('Mean Mission Duration Time')
plt.xlabel('Nationality')
plt.ylabel('Mission Duration (hrs)');
```

```python
plt.figure(figsize=(15,7))
astronauts_df['total_hrs_sum'] = astronauts_df['total_hrs_sum'].astype(float)
astronauts_df['total_hrs_sum'].groupby(astronauts_df['nationality']).sum().sort_values(ascending = False)[0:10].plot(kind='bar')
plt.title('Mission Duration Time')
plt.xlabel('Nationality')
plt.ylabel('Total Mission Duration (hrs)');
```

```python
top_5_astronauts = astronauts_df.groupby(['name', 'nationality', 'occupation', 'sex'])['total_hrs_sum'].value_counts()[0:5].sort_values(ascending = False).to_frame()
top_5_astronauts = top_5_astronauts.iloc[:,:-1]
top_5_astronauts
```

```python
plt.figure(figsize = (12, 7))
inOrbit = astronauts_df['in_orbit'].value_counts()[0:5]
inOrbit.plot(kind = 'barh', color ='green')
plt.title('Top 5 In Orbit Count');
```

```python
plt.figure(figsize = (12, 7))
inOrbit = astronauts_df['ascend_shuttle'].value_counts()[0:5]
inOrbit.plot(kind = 'barh', color = 'steelblue')
plt.title('Top 5 Ascension Shuttle Count');
```

```python
n_bins=10
plt.figure(figsize=(12,7))
plt.hist(astronauts_df.year_of_mission, n_bins, color = 'purple')
plt.xlabel('Year')
plt.ylabel('Number of missions')
plt.title('Number of missions over time');
```

```python
plt.gcf().set_size_inches(12,7)
plt.title('Age and Gender by Year of Mission')
sns.scatterplot(data = astronauts_df, x = 'age', y = 'year_of_mission', hue = 'sex');
```

```python
eva = astronauts_df.groupby('selection')['total_eva_hrs', 'eva_hrs_mission'].mean().nlargest(10, columns = 'total_eva_hrs')
eva.plot(kind = 'area', figsize = (12,8))
plt.title('Number of Extravehicular Activities by Selection Program');
```

```python
astronauts_df['training_time'] = astronauts_df['year_of_mission'] - astronauts_df['year_of_selection']
sns.lineplot(x = 'age', y = 'training_time', data = astronauts_df, color = 'red')
plt.gcf().set_size_inches(12,7)
plt.title('Astronauts Training Time by Age');
```
